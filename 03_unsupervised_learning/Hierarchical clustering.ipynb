```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
import pandas as pd

eurovision = pd.read_csv("https://assets.datacamp.com/production/repositories/655/datasets/2a1f3ab7bcc76eef1b8e1eb29afbd54c4ebf86f2/eurovision-2016.csv")
eurovision.head()

df = eurovision.pivot(index = 'From country',
                     columns = 'To country',
                     values = 'Jury Points')

samples = df.to_numpy(na_value = 0.0)

mergings = linkage(samples, method = 'complete')

dendrogram(mergings,
          labels = df.index,
          leaf_rotation = 90,
          leaf_font_size = 6)

plt.show()
```
